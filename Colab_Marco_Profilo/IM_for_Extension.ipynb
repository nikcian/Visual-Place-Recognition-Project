```python
import os
import torch
import pandas as pd
import numpy as np
from pathlib import Path
import re

def get_distances_fixed(txt_file_path):
    distances = []
    if not os.path.exists(txt_file_path): return []
    with open(txt_file_path, "r") as f:
        lines = f.readlines()

    def extract_lat_lon(line):
        # Cerchiamo tutte le coppie @numero@numero@
        matches = re.findall(r"@([-+]?\d+\.\d+)@([-+]?\d+\.\d+)@", line)
        if len(matches) >= 2:
            # Prendiamo la SECONDA coppia (quella con 37.xxx e -122.xxx)
            return float(matches[1][0]), float(matches[1][1])
        elif len(matches) == 1:
            return float(matches[0][0]), float(matches[0][1])
        return None

    query_coords = extract_lat_lon(lines[1])
    if query_coords is None: return []

    for line in lines[3:]:
        pred_coords = extract_lat_lon(line)
        if pred_coords:
            # Formula Haversine corretta
            lat1, lon1 = np.radians(query_coords)
            lat2, lon2 = np.radians(pred_coords)
            dlat, dlon = lat2 - lat1, lon2 - lon1
            a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
            dist = 6371000 * (2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)))
            distances.append(dist)
    return distances

# --- CONFIGURAZIONE ---
matcher = "loftr"
split = "test"
inliers_local = f"/content/Visual-Place-Recognition-Project/preds_{split}_txt_{matcher}"
preds_txt_dir = f"/content/Visual-Place-Recognition-Project/preds_{split}_txt"
drive_csv_path = f"/content/drive/MyDrive/VPR_6.1_Results/sf_xs/{split}/CSVs/{matcher}_{split}_report_6_1.csv"

# --- ESECUZIONE ---
print(f"ğŸ”„ Rigenerazione CSV con coordinate GPS corrette...")
if os.path.exists(inliers_local):
    data_list = []
    torch_files = sorted(list(Path(inliers_local).glob("*.torch")), key=lambda x: int(x.stem))

    for t_file in torch_files:
        geo_dists = get_distances_fixed(os.path.join(preds_txt_dir, f"{t_file.stem}.txt"))
        if not geo_dists: continue

        results = torch.load(t_file, weights_only=False)
        inliers = [r['num_inliers'] for r in results]
        max_inliers = max(inliers) if inliers else 0
        best_idx = inliers.index(max_inliers) if inliers else 0

        # Verifica se entro 25 metri
        is_correct = bool(geo_dists[best_idx] <= 25)

        data_list.append({'query_id': t_file.stem, 'max_inliers': max_inliers, 'is_correct': is_correct})

    df = pd.DataFrame(data_list)
    os.makedirs(os.path.dirname(drive_csv_path), exist_ok=True)
    df.to_csv(drive_csv_path, index=False)

    print(f"âœ… CSV SALVATO: {drive_csv_path}")
    print(f"ğŸ“Š Statistiche: {df['is_correct'].value_counts().to_dict()}")
    print("\nğŸ‘€ Esempi con inliers alti:")
    print(df[df['max_inliers'] > 100].head())
```

```python
import os
import torch
import pandas as pd
import numpy as np
from pathlib import Path
import re

# --- FUNZIONE DI ESTRAZIONE COORDINATE CORRETTE ---
def get_distances_fixed_lg(txt_file_path):
    distances = []
    if not os.path.exists(txt_file_path): return []
    with open(txt_file_path, "r") as f:
        lines = f.readlines()

    def extract_lat_lon(line):
        # Cerchiamo le coppie @lat@lon@.
        # Prendiamo la SECONDA coppia matches[1] (quella 37.xxx, -122.xxx)
        matches = re.findall(r"@([-+]?\d+\.\d+)@([-+]?\d+\.\d+)@", line)
        if len(matches) >= 2:
            return float(matches[1][0]), float(matches[1][1])
        elif len(matches) == 1:
            return float(matches[0][0]), float(matches[0][1])
        return None

    query_coords = extract_lat_lon(lines[1])
    if query_coords is None: return []

    for line in lines[3:]:
        pred_coords = extract_lat_lon(line)
        if pred_coords:
            lat1, lon1 = np.radians(query_coords)
            lat2, lon2 = np.radians(pred_coords)
            dlat, dlon = lat2 - lat1, lon2 - lon1
            a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
            dist = 6371000 * (2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)))
            distances.append(dist)
    return distances

# --- CONFIGURAZIONE ---
matcher = "superpoint-lg"
split = "test"
inliers_local = f"/content/Visual-Place-Recognition-Project/preds_{split}_txt_{matcher}"
preds_txt_dir = f"/content/Visual-Place-Recognition-Project/preds_{split}_txt"
drive_csv_path = f"/content/drive/MyDrive/VPR_6.1_Results/sf_xs/{split}/CSVs/{matcher}_{split}_report_6_1.csv"

# --- ESECUZIONE ---
print(f"ğŸ› ï¸ Riparazione CSV per {matcher.upper()} {split.upper()} con Fix Coordinate...")

if os.path.exists(inliers_local):
    data_list = []
    torch_files = sorted(list(Path(inliers_local).glob("*.torch")), key=lambda x: int(x.stem))

    for t_file in torch_files:
        txt_path = os.path.join(preds_txt_dir, f"{t_file.stem}.txt")
        geo_dists = get_distances_fixed_lg(txt_path)
        if not geo_dists: continue

        results = torch.load(t_file, weights_only=False)
        inliers = [r['num_inliers'] for r in results]

        # Troviamo il migliore basato sugli inliers (Reranking locale)
        max_inliers = max(inliers) if inliers else 0
        best_idx = inliers.index(max_inliers) if inliers else 0

        # Verifichiamo se la distanza reale Ã¨ <= 25 metri
        is_correct = bool(geo_dists[best_idx] <= 25)

        data_list.append({
            'query_id': t_file.stem,
            'max_inliers': max_inliers,
            'is_correct': is_correct
        })

    # Salvataggio su Drive
    df = pd.DataFrame(data_list)
    os.makedirs(os.path.dirname(drive_csv_path), exist_ok=True)
    df.to_csv(drive_csv_path, index=False)

    print(f"âœ… CSV SISTEMATO: {drive_csv_path}")
    print(f"ğŸ“Š Statistiche True/False: {df['is_correct'].value_counts().to_dict()}")
    print(df.head(5))
else:
    print(f"âŒ Errore: Cartella locale {inliers_local} non trovata.")
```

```python
import os
import shutil
import torch
import pandas as pd
import numpy as np
from pathlib import Path
import re

# --- 1. CONFIGURAZIONE PERCORSI ---
matcher = "superpoint-lg" # <--- CAMBIATO IN LIGHTGLUE
split = "test"
query_type = "queries_sun"

# Root del progetto
project_root = "/content/Visual-Place-Recognition-Project"
script_path = f"{project_root}/match_queries_preds.py"

# Cartella dei TXT (giÃ  esistente dal test con LoFTR)
txt_dir_svox = f"{project_root}/svox_{split}_{query_type}_txt"
inliers_local = f"{txt_dir_svox}_{matcher}"

# Destinazioni su Drive
drive_root = f"/content/drive/MyDrive/VPR_6.1_Results/svox/{split}/{query_type}"
drive_dest_torch = f"{drive_root}/torch_files/{matcher}"
drive_dest_csv = f"{drive_root}/CSVs"
csv_filename = f"{matcher}_{query_type}_report_6_1.csv"

# --- 2. ESECUZIONE MATCHING ---
if os.path.exists(script_path):
    print(f"ğŸš€ Avvio Matching {matcher.upper()}...")
    %cd {project_root}

    !python match_queries_preds.py \
        --preds-dir "{txt_dir_svox}" \
        --matcher {matcher} \
        --device cuda \
        --num-preds 20
else:
    print(f"âŒ ERRORE: Script non trovato in {script_path}")

# --- 3. FUNZIONE DISTANZA UTM (SVOX) ---
def get_distances_svox(txt_file_path):
    distances = []
    if not os.path.exists(txt_file_path): return []
    with open(txt_file_path, "r") as f:
        lines = f.readlines()
    def extract_utm(line):
        matches = re.findall(r"@([-+]?\d+\.\d+)@([-+]?\d+\.\d+)@", line)
        if matches: return float(matches[0][0]), float(matches[0][1])
        return None
    query_coords = extract_utm(lines[1])
    if query_coords is None: return []
    for line in lines[3:]:
        pred_coords = extract_utm(line)
        if pred_coords:
            dist = np.sqrt((query_coords[0] - pred_coords[0])**2 + (query_coords[1] - pred_coords[1])**2)
            distances.append(dist)
    return distances

# --- 4. BACKUP E GENERAZIONE CSV 6.1 ---
if os.path.exists(inliers_local):
    print(f"\nğŸ“¦ Backup .torch su Drive...")
    os.makedirs(os.path.dirname(drive_dest_torch), exist_ok=True)
    if os.path.exists(drive_dest_torch): shutil.rmtree(drive_dest_torch)
    shutil.copytree(inliers_local, drive_dest_torch)

    print(f"ğŸ“„ Generazione Report CSV...")
    torch_files = sorted(list(Path(inliers_local).glob("*.torch")), key=lambda x: int(x.stem))
    data_list = []

    for t_file in torch_files:
        txt_path = os.path.join(txt_dir_svox, f"{t_file.stem}.txt")
        geo_dists = get_distances_svox(txt_path)
        if not geo_dists: continue

        results = torch.load(t_file, weights_only=False)
        inliers = [r['num_inliers'] for r in results]

        if inliers:
            max_inliers = max(inliers)
            best_idx = inliers.index(max_inliers)
            is_correct = 1 if geo_dists[best_idx] <= 25 else 0

            data_list.append({
                'query_id': t_file.stem,
                'max_inliers': max_inliers,
                'is_correct': is_correct
            })

    os.makedirs(drive_dest_csv, exist_ok=True)
    df = pd.DataFrame(data_list)
    df.to_csv(f"{drive_dest_csv}/{csv_filename}", index=False)
    print(f"âœ… CSV SALVATO SU DRIVE: {drive_dest_csv}/{csv_filename}")
    print(f"ğŸ“Š Statistiche {matcher}:")
    print(df['is_correct'].value_counts())

    # --- 5. RERANKING FINALE ---
    print(f"\nğŸ“ˆ Calcolo Recall finale...")
    !python reranking.py --preds-dir "{txt_dir_svox}" --inliers-dir "{inliers_local}" --num-preds 20
else:
    print(f"âŒ ERRORE: La cartella {inliers_local} non Ã¨ stata creata.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- CONFIGURAZIONE PERCORSI ---
# Inserisci qui i percorsi ai tuoi nuovi file CSV su Drive
csv_files = {
    "SVOX Sun - LoFTR": "/content/drive/MyDrive/VPR_6.1_Results/svox/test/queries_sun/CSVs/loftr_queries_sun_report_6_1.csv",
    "SVOX Sun - SP+LG": "/content/drive/MyDrive/VPR_6.1_Results/svox/test/queries_sun/CSVs/superpoint-lg_queries_sun_report_6_1.csv",
    "SVOX Night - LoFTR": "/content/drive/MyDrive/VPR_6.1_Results/svox/test/queries_night/CSVs/loftr_queries_night_report_6_1.csv",
    "SVOX Night - SP+LG": "/content/drive/MyDrive/VPR_6.1_Results/svox/test/queries_night/CSVs/superpoint-lg_queries_night_report_6_1.csv"
}

def plot_inliers_distribution(csv_path, title):
    if not os.path.exists(csv_path):
        print(f"âš ï¸ File non trovato: {title}")
        return

    df = pd.read_csv(csv_path)

    # Separazione corretti (1) e sbagliati (0)
    correct = df[df['is_correct'] == 1]['max_inliers']
    wrong = df[df['is_correct'] == 0]['max_inliers']

    plt.figure(figsize=(12, 7))

    # Plot istogrammi
    plt.hist(correct, bins=50, alpha=0.6, color='green', label='Correct Queries (R@1)', edgecolor='black')
    plt.hist(wrong, bins=50, alpha=0.6, color='red', label='Wrong Queries', edgecolor='black')

    # Configurazione estetica
    plt.title(f"Distribution of Inliers for Correct vs Wrong Queries\n({title})", fontsize=15)
    plt.xlabel(f"Number of Inliers ({'LightGlue' if 'SP+LG' in title else 'LoFTR'})", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.legend(loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

# --- GENERAZIONE GRAFICI ---
for title, path in csv_files.items():
    plot_inliers_distribution(path, title)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from sklearn.metrics import precision_recall_curve, auc

# --- 1. CONFIGURAZIONE PERCORSI ---
# Nuovo percorso base come indicato
drive_base = "/content/drive/MyDrive/NETVLAD_DATI/VPR_Project_Final_Results/svox/test"
output_dir = "svox_test_plots_final"
os.makedirs(output_dir, exist_ok=True)

# Dataset e Matcher
datasets = ["queries_sun", "queries_night"]
matchers = ["loftr", "superpoint-lg"]

def generate_vpr_plots(csv_path, dataset_name, matcher_name):
    if not os.path.exists(csv_path):
        print(f"âš ï¸ Salto: {csv_path} non trovato.")
        return None

    df = pd.read_csv(csv_path)
    # Assicurati che le colonne si chiamino correttamente nel tuo CSV
    inlier_col = 'max_inliers'
    correct_col = 'is_correct'

    print(f"ğŸ“ˆ Elaborazione: {dataset_name} + {matcher_name}")

    # --- 2. PLOT 6.1: ADAPTIVE RE-RANKING ---
    max_val = int(df[inlier_col].max())
    thresholds = np.linspace(0, max_val, 100)
    recalls = []
    cost_savings = []
    total_q = len(df)

    for t in thresholds:
        correct_above_t = df[(df[correct_col] == 1) & (df[inlier_col] >= t)].shape[0]
        recalls.append((correct_above_t / total_q) * 100)

        easy_queries = df[df[inlier_col] >= t].shape[0]
        cost_savings.append((easy_queries / total_q) * 100)

    fig, ax1 = plt.subplots(figsize=(10, 5))
    ax1.plot(thresholds, recalls, color='tab:blue', label='Recall@1 (Adaptive)', linewidth=2)
    ax1.set_xlabel('Inlier Threshold (Ï„)')
    ax1.set_ylabel('Recall@1 (%)', color='tab:blue')

    ax2 = ax1.twinx()
    ax2.plot(thresholds, cost_savings, color='tab:green', linestyle='--', label='Cost Saving (%)')
    ax2.set_ylabel('Potential Computational Saving (%)', color='tab:green')

    plt.title(f'6.1 Adaptive Re-ranking: SVOX {dataset_name} ({matcher_name})')
    plt.grid(alpha=0.3)
    plt.savefig(f"{output_dir}/svox_{dataset_name}_{matcher_name}_6.1_adaptive.png")
    plt.close()

    # --- 3. PLOT 6.2: UNCERTAINTY (AUPRC) ---
    uncertainty_score = -df[inlier_col]
    y_true = 1 - df[correct_col]

    precision, recall, _ = precision_recall_curve(y_true, uncertainty_score)
    auprc_score = auc(recall, precision)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='darkorange', label=f'PR Curve (AUPRC = {auprc_score:.3f})')
    plt.xlabel('Recall (Detection of Errors)')
    plt.ylabel('Precision (Confidence in Error Prediction)')
    plt.title(f'6.2 Uncertainty Estimation: SVOX {dataset_name} ({matcher_name})')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.savefig(f"{output_dir}/svox_{dataset_name}_{matcher_name}_6.2_auprc.png")
    plt.close()

    return auprc_score

# --- 4. ESECUZIONE LOOP ---
summary = []
for ds in datasets:
    for mt in matchers:
        # Costruzione percorso: drive_base/queries_sun/CSVs/loftr_queries_sun.csv
        csv_p = f"{drive_base}/{ds}/CSVs/{mt}_{ds}.csv"

        # Correzione specifica per il nome del file se necessario
        # (alcuni sistemi salvano superpoint-lg come superpoint_lg)
        if not os.path.exists(csv_p) and mt == "superpoint-lg":
            csv_p = f"{drive_base}/{ds}/CSVs/superpoint_lg_{ds}.csv"

        score = generate_vpr_plots(csv_p, ds, mt)
        if score:
            summary.append({'Dataset': ds, 'Matcher': mt, 'AUPRC': score})

# --- 5. TABELLA RIASSUNTIVA ---
if summary:
    df_summary = pd.DataFrame(summary)
    print("\n--- NUOVI RISULTATI AUPRC (SVOX TEST) ---")
    print(df_summary)
    df_summary.to_csv(f"{output_dir}/svox_summary_metrics.csv", index=False)
else:
    print("âŒ Nessun file CSV trovato. Controlla che i nomi dei file su Drive siano esattamente 'loftr_queries_sun.csv', ecc.")
```
